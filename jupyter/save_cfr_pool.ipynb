```python
import os
import pandas as pd

model_dir = '/Users/morton/Dropbox/work/cmr_cfr/cfr'
data_file = os.path.join(model_dir, 'cfr_eeg_mixed.csv')
data_dir = '/Users/morton/PycharmProjects/cymr/src/cymr/data'
```

```python
data = pd.read_csv(data_file, dtype={'category': 'category'})
data.category.cat.as_ordered(inplace=True)
data.head()
data.to_csv(os.path.join(data_dir, 'Morton2013.csv'), index=False)
```

```python
pool = data.groupby('item_index')[['item', 'category']].first()
pool.index = pool.index.astype('int').rename('item_index')
pool.to_csv(os.path.join(data_dir, 'Morton2013_pool.csv'))
```
